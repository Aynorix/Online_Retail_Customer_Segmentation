```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
df = pd.read_csv("retail_sales_dataset (1).csv")

#converts categorical columns
df = pd.get_dummies(df, columns=["Gender", "Product Category"], drop_first=True)

X = df.drop(["Transaction ID","Date","Customer ID","Total Amount"], axis=1)
y = df["Total Amount"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("R squared score:", r2_score(y_test, y_pred))

```

```python
# matplotlib
plt.figure(figsize=(13,3))
plt.plot(y_test.values, label="Actual Total Amount")
plt.plot(y_pred, label="Predicted Total Amount", color='r')

plt.xlabel("Predicted Amount")
plt.ylabel("Total Amount")
plt.title("Actual vs Predicted Total Amount")

plt.legend()
plt.show()
```
